```python
df.Date=pd.to_datetime(df.Date)
```

```python
len(df.Arrest)
```

```python
df['year']=df.Date.dt.year
df['day']=df.Date.dt.weekday_name
df['month']=df.Date.dt.month
```

```python
df.groupby([df.month]).size().plot.bar()
plt.title('Crime Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes Acts')
plt.show()
```

```python
df.groupby([df.day]).size().plot.bar()
plt.title('Crime Per day')
plt.xlabel('day')
plt.ylabel('Number of Crimes Acts')
plt.show()
```

```python
df.groupby([df.year]).size().plot.bar()
plt.title('Crime Per year')
plt.xlabel('year')
plt.ylabel('Number of Crimes Acts')
plt.show()
```
